### Move a specific file to specific folder
suppose you have a file image.pdf. You wan to move the file to a folder called "folder".

```python
import shutil

filename = 'C:/Users/Asus/Desktop/image.pdf'

target = r'C:/Users/Asus/Desktop/folder/'

shutil.move(filename, target)
```

### Read filenames from a csv file and move to specific folder
Suppose you have many files in a directory. And you have a csv that contains only the names of the files that you want to move. So using the code below you can move those specific files.


```python

import csv
import os
import shutil
Csv_File = 'C:/Users/Asus/Desktop/465_Dataset/Noncovidlables.csv'
i = 0
count = 0
# opening the CSV file
with open(Csv_File, mode ='r')as file:
   
  # reading the CSV file
  csvFile = csv.reader(file)
     
  print(csvFile)
  for filename in csvFile:
        filename = str(filename[i])
        original_directory = 'C:/Users/Asus/Desktop/images/'
        target = r'C:/Users/Asus/Desktop/465_Dataset/Noncovid/'
        if os.path.exists(original_directory+filename):
                shutil.move(original_directory+filename, target)
        else:
                print(filename)
                count = count + 1

print(count)
        
             
              
        
        
        

        
        
         
```
